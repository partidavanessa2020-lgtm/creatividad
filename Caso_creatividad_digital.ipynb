```python
#Indicadores basicos
df['ctr'] = df['clicks'] / df['impresiones'] #Tasa de clicks
df['cvr'] = df['conversiones'] / df['clicks'] #Tasa de conversion

#Indicadores de costo
df['cpc'] = df['costo_usd'] / df['clicks'] #Costo por click
df['cpm'] = (df['costo_usd'] / df['impresiones']) * 1000 #Costo por miles

#Indicador compuesto
df['roi'] = (df['ingresos_usd'] - df['costo_usd']) / df['costo_usd'] #Retorno sobre inversion
```
